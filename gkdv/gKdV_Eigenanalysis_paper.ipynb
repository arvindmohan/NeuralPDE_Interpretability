```python
import numpy as np
import os
import h5py
import matplotlib.pyplot as plt

def nth_root(x, n):
    return np.power(x, 1/n)

fpath_expt1 = "./Expt1/jacobian_gKdV_caseA_dict.npz"
fpath_expt2 = "./Expt2/jacobian_gKdV_caseB_dict.npz"
print(os.getcwd())
c_list = [2.5,1.5,2,3,3.5,4,4.5,5]

#def unit circle
theta = np.arange(-np.pi,np.pi,0.01)
x=np.cos(theta)
y=np.sin(theta)

# Expt 1
data1 = np.load(fpath_expt1, allow_pickle=True)
jac_true_all_expt1 = np.asarray(data1["jac_true_all"]);
jac_pred_all_expt1 = np.asarray(data1["jac_pred_all"]);
# Expt 2
data2 = np.load(fpath_expt2, allow_pickle=True)
jac_true_all_expt2 = np.asarray(data2["jac_true_all"]);
jac_pred_all_expt2 = np.asarray(data2["jac_pred_all"]);
```

## Study eigenvalues for IC inside training set and only 1 IC outside training set.
For brevity. Can make separate plot for all test ICs later.

```python
# select jacobians
jac_true_c0_expt1 = jac_true_all_expt1[:,:,0] #c=2.5 training
jac_true_c15_expt1 = jac_true_all_expt1[:,:,1]
jac_true_c2_expt1 = jac_true_all_expt1[:,:,2]
jac_true_c3_expt1 = jac_true_all_expt1[:,:,3]
jac_true_c35_expt1 = jac_true_all_expt1[:,:,4]
jac_true_c4_expt1 = jac_true_all_expt1[:,:,5]
jac_true_c45_expt1 = jac_true_all_expt1[:,:,6]
jac_true_c5_expt1 = jac_true_all_expt1[:,:,7]

jac_pred_c0_expt1 = jac_pred_all_expt1[:,:,0]
jac_pred_c15_expt1 = jac_pred_all_expt1[:,:,1]
jac_pred_c2_expt1 = jac_pred_all_expt1[:,:,2]
jac_pred_c3_expt1 = jac_pred_all_expt1[:,:,3]
jac_pred_c35_expt1 = jac_pred_all_expt1[:,:,4]
jac_pred_c4_expt1 = jac_pred_all_expt1[:,:,5]
jac_pred_c45_expt1 = jac_pred_all_expt1[:,:,6]
jac_pred_c5_expt1 = jac_pred_all_expt1[:,:,7]

jac_pred_c0_expt2 = jac_pred_all_expt2[:,:,0]
jac_pred_c15_expt2 = jac_pred_all_expt2[:,:,1]
jac_pred_c2_expt2 = jac_pred_all_expt2[:,:,2]
jac_pred_c3_expt2 = jac_pred_all_expt2[:,:,3]
jac_pred_c35_expt2 = jac_pred_all_expt2[:,:,4]
jac_pred_c4_expt2 = jac_pred_all_expt2[:,:,5]
jac_pred_c45_expt2 = jac_pred_all_expt2[:,:,6]
jac_pred_c5_expt2 = jac_pred_all_expt2[:,:,7]

#compute Eigenvals for true PDE
[etrue_c0_expt1,_]=np.linalg.eig(jac_true_c0_expt1)
[etrue_c15_expt1,_]=np.linalg.eig(jac_true_c15_expt1)
[etrue_c2_expt1,_]=np.linalg.eig(jac_true_c2_expt1)
[etrue_c3_expt1,_]=np.linalg.eig(jac_true_c3_expt1)
[etrue_c35_expt1,_]=np.linalg.eig(jac_true_c35_expt1)
[etrue_c4_expt1,_]=np.linalg.eig(jac_true_c4_expt1)
[etrue_c45_expt1,_]=np.linalg.eig(jac_true_c45_expt1)
[etrue_c5_expt1,_]=np.linalg.eig(jac_true_c5_expt1)

#compute Eigenvals for Expt 1
[epred_c0_expt1, _] = np.linalg.eig(jac_pred_c0_expt1)
[epred_c15_expt1, _] = np.linalg.eig(jac_pred_c15_expt1)
[epred_c2_expt1, _] = np.linalg.eig(jac_pred_c2_expt1)
[epred_c3_expt1, _] = np.linalg.eig(jac_pred_c3_expt1)
[epred_c35_expt1, _] = np.linalg.eig(jac_pred_c35_expt1)
[epred_c4_expt1, _] = np.linalg.eig(jac_pred_c4_expt1)
[epred_c45_expt1, _] = np.linalg.eig(jac_pred_c45_expt1)
[epred_c5_expt1, _] = np.linalg.eig(jac_pred_c5_expt1)

#compute Eigenvals for Expt 2
[epred_c0_expt2, _] = np.linalg.eig(jac_pred_c0_expt2)
[epred_c15_expt2, _] = np.linalg.eig(jac_pred_c15_expt2)
[epred_c2_expt2, _] = np.linalg.eig(jac_pred_c2_expt2)
[epred_c3_expt2, _] = np.linalg.eig(jac_pred_c3_expt2)
[epred_c35_expt2, _] = np.linalg.eig(jac_pred_c35_expt2)
[epred_c4_expt2, _] = np.linalg.eig(jac_pred_c4_expt2)
[epred_c45_expt2, _] = np.linalg.eig(jac_pred_c45_expt2)
[epred_c5_expt2, _] = np.linalg.eig(jac_pred_c5_expt2)

#compute max eigenvalues for true PDE
true_maxeigen_IC = np.zeros([8])
true_maxeigen_IC[0] = np.max(np.abs(etrue_c0_expt1))
true_maxeigen_IC[1] = np.max(np.abs(etrue_c15_expt1))
true_maxeigen_IC[2] = np.max(np.abs(etrue_c2_expt1))
true_maxeigen_IC[3] = np.max(np.abs(etrue_c3_expt1))
true_maxeigen_IC[4] = np.max(np.abs(etrue_c35_expt1))
true_maxeigen_IC[5] = np.max(np.abs(etrue_c4_expt1))
true_maxeigen_IC[6] = np.max(np.abs(etrue_c45_expt1))
true_maxeigen_IC[7] = np.max(np.abs(etrue_c5_expt1))

#compute max Eigenvals for Expt 1
expt1_maxeigen_IC = np.zeros([8])
expt1_maxeigen_IC[0] = np.max(np.abs(epred_c0_expt1))
expt1_maxeigen_IC[1] = np.max(np.abs(epred_c15_expt1))
expt1_maxeigen_IC[2] = np.max(np.abs(epred_c2_expt1))
expt1_maxeigen_IC[3] = np.max(np.abs(epred_c3_expt1))
expt1_maxeigen_IC[4] = np.max(np.abs(epred_c35_expt1))
expt1_maxeigen_IC[5] = np.max(np.abs(epred_c4_expt1))
expt1_maxeigen_IC[6] = np.max(np.abs(epred_c45_expt1))
expt1_maxeigen_IC[7] = np.max(np.abs(epred_c5_expt1))

#compute max Eigenvals for Expt 2
expt2_maxeigen_IC = np.zeros([8])
expt2_maxeigen_IC[0] = np.max(np.abs(epred_c0_expt2))
expt2_maxeigen_IC[1] = np.max(np.abs(epred_c15_expt2))
expt2_maxeigen_IC[2] = np.max(np.abs(epred_c2_expt2))
expt2_maxeigen_IC[3] = np.max(np.abs(epred_c3_expt2))
expt2_maxeigen_IC[4] = np.max(np.abs(epred_c35_expt2))
expt2_maxeigen_IC[5] = np.max(np.abs(epred_c4_expt2))
expt2_maxeigen_IC[6] = np.max(np.abs(epred_c45_expt2))
expt2_maxeigen_IC[7] = np.max(np.abs(epred_c5_expt2))
```

### Analysis on IC inside vs outside Train Set

```python
plt.rcParams['figure.figsize'] = [12, 18]  # Adjusted for 3 subplots
fig, ax = plt.subplots(3, 1)

# 1st subplot c=2.5
ax[0].plot(np.real(etrue_c0_expt1), np.imag(etrue_c0_expt1), 'co', label=r'True PDE $c=2.5$')
ax[0].plot(np.real(epred_c0_expt1), np.imag(epred_c0_expt1), 'b+', label=r'Expt 1 Train $c=2.5$')
ax[0].plot(np.real(epred_c0_expt2), np.imag(epred_c0_expt2), 'k*', label=r'Expt 2 Train $c=2.5$')
ax[0].set_xlabel(r'$Re(\lambda)$', fontsize=18)
ax[0].set_ylabel(r'$Im(\lambda)$', fontsize=18)
ax[0].xaxis.set_tick_params(labelsize=15)
ax[0].yaxis.set_tick_params(labelsize=15)
ax[0].set_xlim(-300, 300)  # Set x-axis limits
ax[0].legend()

# 2nd subplot: c = 4.5
ax[1].plot(np.real(etrue_c45_expt1), np.imag(etrue_c45_expt1), 'co', label=r'True PDE $c=4.5$')
ax[1].plot(np.real(epred_c45_expt1), np.imag(epred_c45_expt1), 'b+', label=r'Expt 1 Test $c=4.5$')
ax[1].plot(np.real(epred_c45_expt2), np.imag(epred_c45_expt2), 'k*', label=r'Expt 2 Test $c=4.5$')
ax[1].set_xlabel(r'$Re(\lambda)$', fontsize=18)
ax[1].set_ylabel(r'$Im(\lambda)$', fontsize=18)
ax[1].xaxis.set_tick_params(labelsize=15)
ax[1].yaxis.set_tick_params(labelsize=15)
ax[1].set_xlim(-300, 300)  # Set x-axis limits
ax[1].legend()

# Third subplot: c = 5
ax[2].plot(np.real(etrue_c5_expt1), np.imag(etrue_c5_expt1), 'co', label=r'True PDE $c=5$')
ax[2].plot(np.real(epred_c5_expt1), np.imag(epred_c5_expt1), 'b+', label=r'Expt 1 Test $c=5$')
ax[2].plot(np.real(epred_c5_expt2), np.imag(epred_c5_expt2), 'k*', label=r'Expt 2 Test $c=5$')
ax[2].set_xlabel(r'$Re(\lambda)$', fontsize=18)
ax[2].set_ylabel(r'$Im(\lambda)$', fontsize=18)
ax[2].xaxis.set_tick_params(labelsize=15)
ax[2].yaxis.set_tick_params(labelsize=15)
ax[2].set_xlim(-300, 300)  # Set x-axis limits
ax[2].legend()

ax[0].title.set_text('(a) Train Inference: Expt 1 vs Expt 2 at $c=2.5$')
ax[1].title.set_text('(b) Test Inference: Expt 1 vs Expt 2 at $c=4.5$')
ax[2].title.set_text('(c) Test Inference: Expt 1 vs Expt 2 at $c=5.0$')

# plt.tight_layout()
# plt.show()

plt.savefig("gKdV_eigenstability_train_test.png", format='png')
```
